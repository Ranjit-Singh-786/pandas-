## missing value handling

```python
import pandas as pd
```

```python
b = {'A':[2,5,4,8,7,6,3,2,5,7,None,8],
     'B':[2,5,4,3,None,6,3,2,5,5,None,8],
    'C':[2,5,4,8,None,6,3,2,5,6,None,8],
    'D':[2,None,4,8,None,6,3,2,5,1,None,9],
    'E':[2,5,4,8,7,6,3,2,5,7,11,8]}
b
```

```python
null_df = pd.DataFrame(b)
null_df
```

```python
null_df.isnull().sum()
```

```python
null_df.fillna(0)     # fill by constant
```

```python
null_df.fillna(method='ffill')       # forward fill
```

```python
null_df.fillna(method='backfill')      # backward fill
```

```python
null_df['A'].fillna(null_df['A'].mean())        # filling by mean
```

```python
null_df.dropna()     # row wise drop  axis = 0
```

```python
null_df.dropna(axis=1)        #axis = 1 column wise
```

```python
null_df.mean()
```

```python
null_df.isnull().sum().sum()
```

```python
null_df['A'].var()
```
