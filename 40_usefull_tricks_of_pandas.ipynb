```python
data['reduce_stop_durationtime']=data['stop_duration'].str.findall('\d+.\d+')
data.head()
```

```python
#exercise for understanding
newlist=[]
for x in data['reduce_stop_durationtime']:
    for i,y in enumerate(x):
        try:
            newlist[i]=y.replace(' ','')
        except:
            newlist.append(y.replace(' ',''))
# enumerate
```
