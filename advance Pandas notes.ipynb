```python
tt=pd.DataFrame(np.arange(1,81).reshape(40,2),columns=['col_1','col_2'])
tt
```

```python
def func(x):
    if x<10:
        return 'G10'
    elif x<50:
        return 'G50'
    elif x<70:
        return 'G70'
    elif x<80:
        return 'G80'
```

```python
current=dt.datetime.now()
tt['col_2'].apply(func)
end=dt.datetime.now()
print(end-current)
```

```python
tt['col_2'].map(func)
end=dt.datetime.now()
print(end-current)
```

```python
tt['col_2']=tt['col_2'].map(func)
end=dt.datetime.now()
print(end-current)
```
